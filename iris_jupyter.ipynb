The next objective was to output a summary of each variable to a single text file

```python
csv_file = "iris.csv"
txt_file = "Iris.txt"
text_list = []

with open(csv_file, "r") as my_input_file:
    for line in my_input_file:
        line = line.split(",", 5) 
        text_list.append(" ".join(line))

with open(txt_file, "w") as my_output_file:
    my_output_file.write("#1\n")
    my_output_file.write("double({},{})\n".format(len(text_list), 5))
    for line in text_list:
        my_output_file.write("  " + line)
    print('File Successfully written.')
```

The above has created a text file that has removed the commas. Using  with statement means that the opened file will close automatically when this block of code completes therefore not blocking any of my downstream processes.

Using the groupby function of pandas I can do statistical analysis of the three classes of iris using the four variables https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html

```python
meanValues= iris_df.groupby('class').mean() #find the means of the four measurements using the groupby  function on class 
maxValues= iris_df.groupby('class').max()
minValues= iris_df.groupby('class').min()


print(meanValues)
print(maxValues)
print(minValues)
```
